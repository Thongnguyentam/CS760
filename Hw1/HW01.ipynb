```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#L_1
x_1 = np.linspace(0, 1, 1000)
x_2 = np.linspace(-1, 0, 1000)
y_1 = 1 - x_1
y_4 = -1 + x_1
y_2 = 1 + x_2
y_3 = -1 - x_2
plt.plot(x_1, y_1, "blue")
plt.plot(x_1, y_4, "blue")
plt.plot(x_2, y_2, "blue")
plt.plot(x_2, y_3, "blue") 
```

```python
#L_inf
inf = np.linspace(-1, 1, 1000)
y_d = np.linspace(-1, -1, 1000)
y_u = np.linspace(1, 1, 1000)
plt.plot(inf, y_d, color = 'blue')
plt.plot(inf, y_u, color = 'blue')
plt.plot(y_d, inf, color = 'blue')
plt.plot(y_u, inf, color = 'blue')
plt.axhline(y = 0, color = 'black')
plt.axvline(x = 0, color = 'black')
```

```python
r = 1 #radius
x = np.linspace(-r, r, 1000)
y = np.sqrt(r-x**2)
plt.figure(figsize = (5,5), dpi = 100)
plt.plot(x,y, color = 'blue')
plt.plot(x,-y, color = 'blue')
plt.axhline(y = 0, color = 'black')
plt.axvline(x = 0, color = 'black')

plt.show()
```

```python

mu = np.array([1,-1]).transpose()
sigma = np.array([[2,0], [0,2]])
np.random.seed(0)
multi = np.random.multivariate_normal(mu, sigma, 100)
```

```python
multi_x = multi[:, 0]
multi_y = multi[:, 1]
```

```python
plt.scatter(multi_x, multi_y)
plt.show()
```

```python
mu1 = np.array([5,0]).transpose()
mu2 = np.array([-5,0]).transpose()
mu = 0.3*mu1 + 0.7*mu2
sigma1 = np.array([[1, 0.25], [0.25, 1]])
sigma2 = np.array([[1, -0.25], [-0.25, 1]])
sigma = 0.3*sigma1**2 + 0.7*sigma2**2 + 0.3*mu1**2 + 0.7*mu2**2 - mu**2
print(sigma)
```

```python
np.random.seed(0)
multi = np.random.multivariate_normal(mu, sigma, 100)
plt.scatter(multi[:, 0], multi[:, 1], s = 0.8)
plt.show()
print(multi)
```
